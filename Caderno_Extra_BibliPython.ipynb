```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
df = pd.read_csv('/content/Walmart.csv')
```

```python
df
```

```python
df.info()
```

# Etapa 1

```python
# Contagem de dias(ID,pessoas) normais e feriados trabalhados. 
df.Holiday_Flag.value_counts()
```

Dias Normais

```python
# Filtrando o DataFrame em dias normais, e depois somando os dias.
df[df['Holiday_Flag']==0].value_counts().sum()
```

```python
# Filtrando o DataFrame em dias normais em uma variável separada.
diasNormais = df[df['Holiday_Flag']==0]
```

```python
# Somando as vendas semais dos dias normais
vendasNormais = diasNormais['Weekly_Sales'].sum()
vendasNormais
```

Dias de Feriado

```python
# Filtrando o DataFrame em feriados, e depois somando os dias.
df[df['Holiday_Flag']==1].value_counts().sum()
```

```python
# Filtrando o DataFrame em feriados em uma variável separada.
diasFeriado = df[df['Holiday_Flag']==1]
```

```python
# Somando as vendas semais dos feriados
vendasFeriado = diasFeriado['Weekly_Sales'].sum()
vendasFeriado
```

```python
df
```

```python
# Mudando o formato de string para data
df['Date'] =  pd.to_datetime(df['Date'], format='%d-%m-%Y')
df
```

```python
df.info()
```

```python
# Criando e adicionado a coluna mês.
df["Month"] = pd.DatetimeIndex(df['Date']).month
```

```python
# Quantidade de vendas feitas em cada mês nos dias normais e feriados
sns.barplot(data=df, x="Month", y="Weekly_Sales",hue='Holiday_Flag');
```

# Etapa 2

```python
# Calculando a média do combustivel em dias normais.
diasNormais.Fuel_Price.mean()
```

```python
# Calculando a média do combustivel em feriados.
diasFeriado.Fuel_Price.mean()
```

```python
# Preço da combustivel em torno dos meses, e também nos dias normais e feriados.
plt.figure(figsize=(10,5))
sns.lineplot(data=df, x = "Month", y="Fuel_Price", hue="Holiday_Flag")
```

```python
# Preço do combustivel em todos os meses ordenado do maior para o menor.
df.groupby(['Month']).agg({'Fuel_Price':'max'}).sort_values('Fuel_Price',ascending = False).reset_index()
```
